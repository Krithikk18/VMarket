```python
import yfinance as yf
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
ticker = ["HDFCBANK.NS","RELIANCE.NS","INFY.NS"]
dataHDFC = yf.download(ticker[0], start = "2015-01-01" , end = "2024-12-31" , interval = "1d")
dataRELIANCE = yf.download(ticker[1], start = "2015-01-01" , end = "2024-12-31" , interval = "1d")
dataINFY = yf.download(ticker[2], start = "2015-01-01" , end = "2024-12-31" , interval = "1d")
data = [dataHDFC,dataRELIANCE,dataINFY]
for i in range (3):
    print(f"{ticker[i]} : {data[i].shape}")
    print(data[i].head())
    print()
```

```python
for i in range(3):
    data[i].columns = data[i].columns.get_level_values(0)
    data[i] = data[i].drop(columns = ["Volume"])
    print(ticker[i])
    print(data[i].head(),"\n")
```

```python
def garman_klass(df):
    Close = df["Close"]
    Open = df["Open"]
    High = df["High"]
    Low = df["Low"]
    GK = 0.5 * (np.log(High/Low))^2 - (2*np.log(2) - 1) * (np.log(Close/Open))^2
    Vol_GK = np.sqrt(GK*252)
```
